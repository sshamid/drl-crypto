## Data Processing

```python
import pandas as pd
import numpy as np
```

```python
df_lstm_out1=pd.read_csv('./data/train_btc.csv')
df_lstm_out2=pd.read_csv('./data/test_btc.csv')
df_lstm_out=pd.concat([df_lstm_out1, df_lstm_out2])

df_lstm_out.Date=pd.to_datetime(df_lstm_out.Date).astype(str)
df_lstm_out['Date'] = df_lstm_out.Date.str.replace('-','').astype(int)
#df_lstm_out['ajexdi'] = 1.0
df_lstm_out.rename(columns={'Date':'datadate','Symbol':'tic','Open':'prcod','High':'prchd',\
                            'Low':'prcld','Close':'prccd','Volume':'cshtrd'}, inplace=True)
#df_lstm_out['tic']='BTC'
df_lstm_out.tail()
```

```python
df_lstm_out.datadate.max(), df_lstm_out.datadate.min()
```

```python
df_eth1=pd.read_csv('./data/train_ETH.csv')
df_eth2=pd.read_csv('./data/test_ETH.csv')
df_eth=pd.concat([df_eth1, df_eth2])

df_eth['Date']=pd.to_datetime(df_eth['Date']).astype(str)
df_eth['Date'] = df_eth.Date.str.replace('-','').astype(int)

#df_eth['ajexdi'] = 1.0
df_eth.rename(columns={'Date':'datadate','Symbol':'tic','Open':'prcod','High':'prchd',\
                            'Low':'prcld','Close':'prccd','Volume':'cshtrd'}, inplace=True)

df_eth.tail()
```

```python
df_eth.datadate.max(), df_eth.datadate.min()
```

```python
cols = ['datadate','tic','prcod','prchd','prcld','prccd','cshtrd']
df=pd.concat([df_lstm_out[df_lstm_out.datadate>=df_eth.datadate.min()][cols], df_eth[df_eth.datadate<=df_lstm_out.datadate.max()][cols]])
```

```python
df.tic.value_counts()
```

```python
df.datadate.max(), df.datadate.min()
```

```python
df_sentiment_score=pd.read_csv('./data/sentiment_score.csv')
df_sentiment_score.date=pd.to_datetime(df_sentiment_score.date).astype(str)
df_sentiment_score.date = df_sentiment_score.date.str.replace('-','').astype(int)
df_sentiment_score.rename(columns={'date':'datadate'}, inplace=True)
df_sentiment_score.head()
```

```python
df2 = df.merge(df_sentiment_score, on='datadate', how='left')
df2['ajexdi'] = 1.0
```
